```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Data Collecting

```python
def load_dataset():
    file = ("merged_data_clean.csv")
    df = pd.read_csv(file)
    return df


load_dataset()
```

```python
def add_manual_data():
    """ 
    Manually add 2019-2021 Google CFE and Grid CFE data from reports.
    """
    manual_data = {
        'Location': ['Singapore', 'Taiwan', 'Belgium', 'Irelad', 'Finland', 'Netherlands',
                     'Chile', 'Iowa', 'SPP - US Southwest', 'Pennsylvania', 'South Carolina	',
                    ],
        '2019 Google CFE (%)': [3, 19, 68, 42, 77, 61, 63, 78, 96, 41, 19],
        '2020 Google CFE (%)': [4, 18, 79, 42, 94, 60, 65, 93, 92, 63, 27],
        '2021 Google CFE (%)': [4, 17, 82, 46, 91, 53, 69, 97, 88, 67, 25],
        '2019 Grid CFE (%)': [3, 19, 68, 42, 76, 24, 42, 29, 41, 41, 19],
        '2020 Grid CFE (%)': [2, 18, 64, 42, 83, 30, 43, 32, 43, 41, 27],
        '2021 Grid CFE (%)': [3, 17, 74, 46, 79, 35, 43, 33, 46, 40, 25]
    }
    
    manual_df = pd.DataFrame(manual_data)
    return manual_df

add_manual_data()
```

```python
def merge_datasets(loaded_data, manual_data):
    """
    Merges the loaded dataset with the manually added data.
    """
    merged_data = pd.merge(loaded_data, manual_data, on='Location', how='outer')
    return merged_data

existing_data = load_dataset()         
manual_data = add_manual_data()
final_data = merge_datasets(existing_data, manual_data)
display(final_data)
```
